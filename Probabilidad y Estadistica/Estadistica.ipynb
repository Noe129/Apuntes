## Descriptiva bivariada / bidimensional
Se le llama asi cuando un individuo está definido por un par de características aleatorias
Una variable es dependiente mientras que la otra independiente 

Ejemplo:
20 estudiantes esta definidos por: chicas (F) y chicos (M). Ademas de si son estudiantes de maestria (MSC) o doctorado(PhD)

```python
Genero = 5 * ["F"] + 15 * ["M"]
Categoria = []
for i in range(20):
    if (np.random.random() > .35):
        Categoria.append("MSc")
    else:
        Categoria.append("PhD")
```

```python
tabla = pd.DataFrame({"Genero": Genero, "Categoria": Categoria})
tabla 
```

### Tabla de doble entrada:
Es cuando se clasifica de a cuerdo a una variable y se ponen separadas las opciones de la otra, en este caso podemos solamente tener fila de F y M y en columnas MSc, PhD

```python
Categoria = "MSc"
ConteoMSc = tabla.query("Categoria==@Categoria")[["Genero"]].value_counts().reset_index()
ConteoMSc.columns = ["Genero", "MSc"]
Categoria = "PhD"
ConteoPhD = tabla.query("Categoria==@Categoria")[["Genero"]].value_counts().reset_index()
ConteoPhD.columns = ["Genero", "PhD"]
ConteoPhD

TabladeDobleEntrada = pd.merge(ConteoPhD, ConteoMSc, on='Genero')
TabladeDobleEntrada
```

```python
TabladeDobleEntrada["Total"] = TabladeDobleEntrada[["PhD", "MSc"]].sum(axis=1)
recuento = TabladeDobleEntrada[["PhD", "MSc", "Total"]].sum(axis=0)
TabladeDobleEntrada = TabladeDobleEntrada.append({"Genero" : "Total", "PhD" : recuento[0], "MSc" : recuento[1], "Total": recuento[2]}, ignore_index = True)
TabladeDobleEntrada
```

```python
plt.style.use('fivethirtyeight')
sm = TabladeDobleEntrada.plot.bar(x="Genero", y=["PhD", "MSc"], rot = 0)
```

# Relacion Entre 2 variables

X e Y estan relacionadas cuando ***por medio de una se puede estimar la otra*** 
Ejemplos:
+ Ingresos y gastos
+ Produccion y venta
+ Altura y peso

Si graficamos los puntos de dispersion, generamos una nube, a al cual le podemos aplicar una recta de regresión
![Recta de Regresion](https://ekuatio.com/wp-content/uploads/recta-de-regresion-12-1.png)

La ***Covarianza*** es la media aritmetica (Promedio) de los productos de las desviaciones de cada una de las variables, esto respecto a las medias.

Como parametro muestral se denota s<sub>XY</sub>

Como parametro poblacional se denota σ<sub>XY</sub>

Si s<sub>XY</sub> es mayor que 0, la relacion es positiva o directa, es decir, la recta de regresion va de abajo para arriba, de izquierda a derecha como se ve en la imagen

La covarianza depende de las unidades, por lo que se usa una medida llamada **Correlacion** r<sub>xy</sub> [-1,1] que se usa para no afectar esto, si es 0 no hay relacion lineal, pero hay otros tipos de relacion. El grado de correlacion depende de que tan cerca estan los puntos de la recta, son correlacion fuerte y debil

$r_{XY} = \frac{S_{XY}}{S_X*S_Y}$


```python
x = [2,3,4,4,5,6,6,7,7,8,10,10]
y = [1,3,2,4,4,4,6,4,6,7,9,10]
multi = lambda x,y : x * y
xpory = list(map(multi, x,y))
cuadrado = lambda x : x * x
xcuadrado = list(map(cuadrado, x))
ycuadrado = list(map(cuadrado, y))
```

```python
NoAgrupados = pd.DataFrame({"X": x, "Y": y, "XporY": xpory, "Xcuadrado": xcuadrado, "Ycuadrado": ycuadrado})
NoAgrupados
```

```python
recuento = NoAgrupados[["X", "Y", "XporY", "Xcuadrado", "Ycuadrado"]].sum(axis=0)
NoAgrupados = NoAgrupados.append({"X" : recuento[0], "Y" : recuento[1], "XporY": recuento[2], "Xcuadrado" : recuento[3], "Ycuadrado": recuento[4]}, ignore_index = True)
lista = list(range(1,13))
lista.append("Total")
NoAgrupados["Indice"] = lista
NoAgrupados.set_index("Indice", inplace = True)
NoAgrupados
```

De aqui pordemos sacar facilmente el promedio de x, de y, la covarianza. El promedio de x es el tota (72) entre el numero de registros, osease 12, lo mismo para la Y.
La covarianza es la suma de xiyi, es decir el total de XporY (431), entre n (12), - la multiplicacion de los promedios

```python
promedioX = np.mean(x)
promedioY = np.mean(y)
Covarianza = (431/12)-(promedioX*promedioY)
Covarianza
```

```python
SigmaX = np.std(x)
SigmaY = np.std(y)
```

```python
correlacion = Covarianza/(SigmaX*SigmaY)
correlacion
```

## Recta de regresion
Ya a la hora de buscarle una aplicación a nuestros datos anteriores, tenemos que lo mas util es calcula la ecuacion de la recta para la recta que nos ayudara a predecir una variable,
Existe la recta de regresion de Y sobre X, la cual estima los valores de Y por medio de X y tambien tenemos la recta de regresion de X sobre Y. La formula de la recta de regresion de y sobre x es la siguiente:

$y-\overline{y} = \frac{S_{XY}}{S_X^2}(x-\overline{x})$

Para la otra solo hay que invertir x con y

```python
x = [2,3,4,4,5,6,6,7,7,8,10,10]
y = [1,3,2,4,4,4,6,4,6,7,9,10]
multi = lambda x,y : x * y
xpory = list(map(multi, x,y))
cuadrado = lambda x : x * x
xcuadrado = list(map(cuadrado, x))
ycuadrado = list(map(cuadrado, y))
NoAgrupados = pd.DataFrame({"X": x, "Y": y, "XporY": xpory, "Xcuadrado": xcuadrado, "Ycuadrado": ycuadrado})
recuento = NoAgrupados[["X", "Y", "XporY", "Xcuadrado", "Ycuadrado"]].sum(axis=0)
NoAgrupados = NoAgrupados.append({"X" : recuento[0], "Y" : recuento[1], "XporY": recuento[2], "Xcuadrado" : recuento[3], "Ycuadrado": recuento[4]}, ignore_index = True)
lista = list(range(1,13))
lista.append("Total")
NoAgrupados["Indice"] = lista
NoAgrupados.set_index("Indice", inplace = True)
NoAgrupados
```

```python
PromedioX = np.mean(x)
PromedioY = np.mean(y)
Covarianza = (431/12)-(PromedioX*PromedioY)
cov = np.cov(x,y, bias=True)
VarianzaX = np.var(x)
VarianzaY = np.var(y)
print(VarianzaX)
print(PromedioX)
print(PromedioY)
print(Covarianza)
print(cov)
NoAgrupados
```
